```python
paths = feature_path+'/txt'
out_path = 'H:/test_data/audio_data/test_audio.csv'
l=[]
for i in range(6373):
    l.append(i)
l = np.array(l).reshape((1,len(l)))
df = pd.DataFrame(l)
for file in os.listdir(paths):
    ID = file.split('.')[0]
    label = 0
    for i in range(audio_label.shape[0]):
        if audio_label.iloc[i][1]==ID:
            label = audio_label.iloc[i][3]
            break
    path = paths+'/'+file
    print(path)
    print(label)
    with open(path, "r", encoding="utf8") as f:
        lines = f.readlines()
        s = lines[6380].split(',')
        s = s[1:-1] 
        s.append(label)
        s = np.array(s).reshape((1,len(s)))
        df1 = pd.DataFrame(s)
        df = pd.concat([df,df1],ignore_index=True)
df = df[1:]
df.to_csv(out_path)
```
