```python
import os
import pandas as pd
import numpy as np

female_train_D_audio_path = 'H:/data/audio/female_train_D_data'
female_train_ND_audio_path = 'H:/data/audio/female_train_ND_data'
male_train_D_audio_path = 'H:/data/audio/male_train_D_data'
male_train_ND_audio_path = 'H:/data/audio/male_train_ND_data'

female_train_D_audio_feature = 'H:/data/audio_data/female_train_D_feature'
female_train_ND_audio_feature = 'H:/data/audio_data/female_train_ND_feature'
male_train_D_audio_feature = 'H:/data/audio_data/male_train_D_feature'
male_train_ND_audio_feature = 'H:/data/audio_data/male_train_ND_feature'

female_train_D_audio_label_path = 'H:/data/audio/female_train_D_audio.csv'
female_train_ND_audio_label_path = 'H:/data/audio/female_train_ND_audio.csv'
male_train_D_audio_label_path = 'H:/data/audio/male_train_D_audio.csv'
male_train_ND_audio_label_path = 'H:/data/audio/male_train_ND_audio.csv'
```

# 标签

```python
female_train_D_audio_label = pd.read_csv(female_train_D_audio_label_path)
female_train_ND_audio_label = pd.read_csv(female_train_ND_audio_label_path)
male_train_D_audio_label = pd.read_csv(male_train_D_audio_label_path)
male_train_ND_audio_label = pd.read_csv(male_train_ND_audio_label_path)
```

# 1.女性抑郁症

```python
for file in os.listdir(female_train_D_audio_path):
    out_path = female_train_D_audio_feature+'/txt/'+file.split('.')[0]+'.txt'
    path = 'cd /d H:/opensmile-3.0-win-x64/bin && SMILExtract.exe -C H:/opensmile-3.0-win-x64/config/compare16/ComParE_2016.conf -I '+ female_train_D_audio_path +'/'+ file+ ' -O '+ out_path
    os.system(path)
```

```python
paths = female_train_D_audio_feature+'/txt'
out_path = 'H:/data/audio_data/female_train_D_feature/female_train_D_audio.csv'
l=[]
for i in range(6373):
    l.append(i)
l = np.array(l).reshape((1,len(l)))
df = pd.DataFrame(l)
for file in os.listdir(paths):
    ID = file.split('.')[0]
    label = 0
    for i in range(female_train_D_audio_label.shape[0]):
        if female_train_D_audio_label.iloc[i][1]==ID:
            label = female_train_D_audio_label.iloc[i][3]
            break
    path = paths+'/'+file
    print(path)
    print(label)
    with open(path, "r", encoding="utf8") as f:
        lines = f.readlines()
        s = lines[6380].split(',')
        s = s[1:-1] 
        s.append(label)
        s = np.array(s).reshape((1,len(s)))
        df1 = pd.DataFrame(s)
        df = pd.concat([df,df1],ignore_index=True)
df = df[1:]
df.to_csv(out_path)
```
